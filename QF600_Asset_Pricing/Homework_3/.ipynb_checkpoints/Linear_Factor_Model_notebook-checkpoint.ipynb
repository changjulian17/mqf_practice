> Briefly explain (in words, without mathematical equations or formulas) the economic significance and pricing implications of each of the three performance ratios (but not α's).
Please submit your results (including relevant tables and graphs) as an Adobe PDF file to Homework 3.

#### Sortino Ratio
Measures expected deviation from target (or benchmark) return.
Economic Significance: Downside risk of investment is risk that realised return on
investment will fall below target return. Sortino Ratio measures outperformance to a benchmark agianst downside risk.
Pricing Implications: Investors might be willing to pay a greater risk premium for investments with a high Sortino ratio, and better return for the level of downside risk.
##### Sharpe Ratio
Focus: Measures risk premium per unit of standard deviation.
Economic Significance: Used for comparing returns on investments against total risk, both systematic and idiosyncratic per market model.
Pricing Implications: A higher Sharpe ratio suggests a more efficient investment, and better return for the level of market risk.
##### Treynor Ratio
Focus: Measures risk premium per unit of market risk.
Economic Significance: Used for comparing returns on investments against systematic (market) risk only and ignore idiosyncratic risk.
Pricing Implications: A higher Treynor ratio suggests a better return for the level of market risk.
