> - Briefly explain (in words, without mathematical equations or formulas) the economic significance and pricing implications of the intercept and slope coefficients.

#### Intercept (Alpha):

Theoretically, Market Model states that there is no intercept. But in practise Alpha Represents the asset's expected return independent of the market's performance.
Indicates the "excess return" over what would be expected based on the asset's risk level.
A positive intercept suggests that the asset has performed better than expected for its risk, while a negative intercept indicates underperformance.
For passive portfolio, represents pricing error relative to CAPM. For active portfolio, represents abnormal mean return after adjusting for exposure to market risk, due to fund manager’s ability to identify underpriced or overpriced assets.

#### Slope (Asset Beta):

Measures amount of exposure to systematic “market” risk, when investing in i’th asset. Market risk premium represents appropriate economic compensation for taking on average amount of exposure to market risk (i.e., corresponding to unit beta). Negative-beta assets can be used to reduce beta of portfolio and provide insurance against unexpected market.downturn

#### Economic Significance
The SML illustrates CAPM pricing formula with expected return of an asset as a function of its systematic risk (beta). Slop is Treynor Ratio (ratio of risk premium to beta). It shows that higher risk should be compensated with higher expected returns.

#### Pricing Implications
The SML reflects market equilibrium, where assets priced above the line are considered undervalued, while those below are overvalued. Therefore The SML can be used to assess whether a stock is fairly priced relative to its beta and expected return, influencing buy/sell decisions.
